# Безполезная ручная реализация

```python
graph = {
        0: [1, 2, 3],
        1: [0, 2, 4],
        2: [0, 1, 3, 4],
        3: [0, 2, 4],
        4: [1, 2, 3, 5],
        5: [4, 6],
        6: [5]
        }
    
edges = []
for node, neighbors in graph.items():
    for neighbor in neighbors:
        if node < neighbor:
            edges.append((node, neighbor))
```

```python
def compute_density(graph, edges):
    n = len(graph)
    m = len(edges)
    return (2 * m) / (n * (n - 1)) if n > 1 else 0.0
```

```python
def compute_clustering_coefficients(graph):
    clustering_local = {}
    for node in graph:
        neighbors = graph[node]
        k = len(neighbors)
        if k < 2:
            clustering_local[node] = 0.0
        else:
            triangles = 0
            for i in range(k):
                for j in range(i + 1, k):
                    if neighbors[j] in graph[neighbors[i]]:
                        triangles += 1
            clustering_local[node] = (2 * triangles) / (k * (k - 1))
    avg_clustering = sum(clustering_local.values()) / len(graph) 
    return clustering_local, avg_clustering
```

```python
def compute_degree_distribution(graph):
    degrees = [len(neighbors) for neighbors in graph.values()]
    total_nodes = len(graph)
    distr = {d:0 for d in degrees}
    for d in degrees:
        distr[d] += 1.0 / total_nodes
    return distr
```

```python
def bfs(graph, start, visited):
    component = []
    queue = deque([start])   
    visited.add(start)
    
    while queue:
        vertex = queue.popleft()
        component.append(vertex)
        
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    
    return component
```

```python
def find_connected_components_bfs(graph):
    visited = set()
    components = []
    
    for vertex in graph:
        if vertex not in visited:
            component = bfs(graph, vertex, visited)
            components.append(component)
    
    return components
```

```python
def compute_closeness_centrality(graph):
    nodes = list(graph.keys())
    n = len(nodes)
    closeness = {}
    for node in nodes:
        dist = bfs(graph, node)
        total_dist = sum(d for d in dist.values() if d > 0)
        closeness[node] = (n - 1) / total_dist
    return closeness
```

```python
def compute_diameter_and_average_distance(graph):
    nodes = list(graph.keys())
    n = len(nodes)
    total_distance = 0
    count = 0
    diameter = 0
    
    for i in range(n):
        dist_i = bfs(graph, nodes[i])
        for j in range(i + 1, n):
            if dist_i[nodes[j]] != -1:
                total_distance += dist_i[nodes[j]]
                count += 1
                diameter = max(diameter, dist_i[nodes[j]])
    
    avg_distance = total_distance / count 
    return diameter, avg_distance
```

```python
# Плотность сети
compute_density(graph, edges)
```

```python
#Локальный, средний коэффициент кластеризациb
compute_clustering_coefficients(graph)
```

```python
#Распределение степеней
compute_degree_distribution(graph)
```

```python
#Поиск компонент связности
find_connected_components_bfs(graph)
```

```python
def bfs(graph, start):
    distances = {node: -1 for node in graph}
    visited = set()
    queue = deque([start])
    visited.add(start)
    distances[start] = 0

    while queue:
        vertex = queue.popleft()
        
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
                distances[neighbor] = distances[vertex] + 1
                
    return distances
```

```python
compute_closeness_centrality(graph)
```

```python
#Диаметр и среднее кратчайшее расстояние
compute_diameter_and_average_distance(graph)
```
